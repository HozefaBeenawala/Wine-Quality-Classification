```python
# Create a boxplot for each column
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed

# Iterate through each column in the dataset
for column in df.columns:
    # Create a boxplot for the current column
    plt.boxplot(df[column], vert=False)
    plt.xlabel(column)  # Set the label for the x-axis
    plt.ylabel('Values')  # Set the label for the y-axis
    plt.show()
```

```python
# Calculate the Z-score for each data point in the dataset
z_scores = np.abs((df - df.mean()) / df.std())

# Define a threshold for outlier detection (e.g., Z-score > 3)
threshold = 3

# Create a new dataset without outliers
df = df[(z_scores < threshold).all(axis=1)]

# Create a boxplot for each column in the new dataset (without outliers)
plt.figure(figsize=(12, 6))  # Adjust the figure size if needed

# Iterate through each column in the new dataset
for column in df.columns:
    # Create a boxplot for the current column
    plt.boxplot(df[column], vert=False)
    plt.xlabel(column)  # Set the label for the x-axis
    plt.ylabel('Values')  # Set the label for the y-axis
    plt.show()  # Show the boxplot
```

```python
#finding correlation between Target variable and Features
df_X = df.drop('quality', axis=1)  # Replace 'target_variable' with the actual name of your target variable column
df_y = df['quality']

# Calculate the correlation matrix
correlation_matrix = pd.concat([df_X, df_y], axis=1).corr()

plt.figure(figsize=(10, 8))  # Adjust the figure size if needed
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
df['good_quality'] = np.where(df['quality'] >= 6, 1, 0)
fig, ax = plt.subplots(figsize=(8, 4))
sns.histplot(data=df, x='quality', hue='good_quality', stat='count', palette=['#929292', '#2ca02c'])
ax.set_title('Wine Quality Score Distribution')
ax.grid(False)
```

```python
df.drop(['Id','quality'],axis=1,inplace=True)
```

```python
df.columns
```

```python
X=df.drop('good_quality',axis=1)
y=df['good_quality']
```

```python
#From Logistic Regression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from sklearn.model_selection import cross_val_score
from sklearn.metrics import recall_score, f1_score, accuracy_score, precision_score

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)

model_lr = LogisticRegression(random_state=3)

model_lr.fit(X_train, y_train)

y_pred = model_lr.predict(X_test)

print(f'The recall_score is : {recall_score(y_test, y_pred)}')  
print(f'The f1_score is : {f1_score(y_test, y_pred)}') 
print(f'The precision_score is : {precision_score(y_test, y_pred)}')  
print(f'The accuracy_score is : {accuracy_score(y_test, y_pred)}\n') 


scores = cross_val_score(model_lr, X_train, y_train, cv=5)
mean_accuracy = scores.mean()

print(f"Mean training accuracy (5-fold CV): {mean_accuracy:.2f}")
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap="Blues")
```
